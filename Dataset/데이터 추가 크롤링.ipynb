### 민간 일자리정보

```python
url = 'http://www.goldenjob.or.kr/job/find-person_view.asp?idx=2457&p=1&keyword=&keyfield='    

def fetch_job_details(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            job_details = {}
            for td in soup.find_all('td'):
                spans = td.find_all('span', class_='mobile')
                if spans:
                    column_name = spans[0].text.strip()  # 컬럼명 추출
                    # 컬럼명 바로 뒤 텍스트 추출(가장 처음의 <span> 태그 다음 텍스트)
                    data_content = td.text.replace(spans[0].text, '').strip()
                    job_details[column_name] = data_content
            return job_details
        else:
            print("Failed to retrieve data: ", response.status_code)
    
    except Exception as e:
        print("An error occurred: ", e)
        
job_data = fetch_job_details(url)

if job_data:
    for key, value in job_data.items():
        print(f"{key}: {value}")
else:
    print("No data found or an error occurred.")
```

```python
data_list = []
```

```python
base_url = 'http://www.goldenjob.or.kr/job/find-person.asp?keyfield=&keyword=&p='

for page in range(1, 163):
    print(f"{page}페이지 크롤링 중 ...")
    page_url = f"{base_url}{page}"
    page_response = requests.get(page_url)
    page_soup = BeautifulSoup(page_response.text, 'html.parser')
    
    # 링크 추출: <a> 태그 내 href 속성에 "find-person_view.asp?idx="가 포함된 모든 링크
    links = page_soup.select('td a[href*="find-person_view.asp?idx="]')
    
    if not links:
        print("No links found on the page. Check the CSS selector.")
        continue

    for link in links:
        # print(f"{link.text.strip()} 크롤링 중 ...")
        detail_url = urljoin(base_url, link['href'])  # 상대 URL을 절대 URL로 변환
        job_info = fetch_job_details(detail_url)
        
        if job_info:
            data_list.append(job_info)
```

```python
df = pd.DataFrame(data_list)
```

### 서울일자리포털

```python
def get_job_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    job_details = {}
    
    # 제목 추출
    title_tag = soup.find('h1', class_='tit_article')
    if title_tag:
        job_details['구인제목'] = title_tag.text.strip()  # 제목을 텍스트로 추출하여 저장
    
    columns = [
        "사업장명", "업종", "사업내용", "근로자수", "주소", 
        "모집직종", "모집인원", "학력", "고용형태", "근무예정지", 
        "직무내용", "경력조건", "필요경력", "급여", "평일 근무시간", 
        "근무형태", "접수마감", "전형방법", "접수방법"
    ]
    
    rows = soup.find_all('tr')
    for row in rows:
        cells = row.find_all('th')
        values = row.find_all('td')
        if cells and values:
            for th, td in zip(cells, values):
                key = th.text.strip()
                if key in columns:
                    value = td.text.strip().replace('\xa0', ' ')
                    job_details[key] = value
    
    return job_details
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import requests
from bs4 import BeautifulSoup

def get_job_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    job_details = {}
    columns = [
        "사업장명", "업종", "사업내용", "근로자수", "주소", 
        "모집직종", "모집인원", "학력", "고용형태", "근무예정지", 
        "직무내용", "경력조건", "필요경력", "급여", "평일 근무시간", 
        "근무형태", "접수마감", "전형방법", "접수방법"
    ]
    
    rows = soup.find_all('tr')
    for row in rows:
        cells = row.find_all('th')
        values = row.find_all('td')
        if cells and values:
            for th, td in zip(cells, values):
                key = th.text.strip()
                if key in columns:
                    value = td.text.strip().replace('\xa0', ' ')
                    job_details[key] = value
    return job_details
```

```python
def setup_driver(url):
    driver = webdriver.Chrome("E:/Downloads/chromedriver-win32/chromedriver.exe")
    driver.get(url)
    return driver
```

```python
def crawl(name, num, url):
    driver = setup_driver(url)
    df = pd.DataFrame()  # 모든 데이터를 저장할 데이터프레임
    wait = WebDriverWait(driver, 20)

    for page in range(1, num+1):  
        print("---------------------------------")
        print(f"Page {page} crawling...")

        # 현재 페이지의 모든 사업장 링크 수집
        wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'td.title_jb a')))
        job_links = [element.get_attribute('href') for element in driver.find_elements(By.CSS_SELECTOR, 'td.title_jb a')]
        current_data = []

        for link in job_links:
            job_details = get_job_details(link)
            current_data.append(job_details)
            print(job_details['구인제목'])

        # 데이터프레임에 추가
        df = pd.concat([df, pd.DataFrame(current_data)], ignore_index=True)

        # 다음 페이지로 이동
        if page < num:
            next_page = page + 1
            driver.execute_script(f"document.jobsearch_frm.pageIndex.value={next_page}; fncGetTbJoList(2);")

        # 중간 저장
        df.to_csv("intermediate_job_data.csv", index=False)

    driver.quit()
    
    print("Crawling completed.")
    df.to_csv(f"{name}_final_job_data.csv", index=False)
    return df
```
